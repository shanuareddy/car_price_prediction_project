Splitting the data into Training data and Test data

```python
X=df.drop(["name","selling_price"],axis=1)
Y=df["selling_price"]
```

```python
print(X)
```

```python
print(Y)
```

Splitting Training and Test Data

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
```

Model Training

1.Linear Regression

```python
#loading the linear regression model 
lin_reg_model=LinearRegression()
```

```python
lin_reg_model.fit(x_train,y_train)
```

```python
#model Evaluvation
#prediction on training data 
pred=lin_reg_model.predict(x_train)
```

```python
# R squared Error
error_score=metrics.r2_score(y_train,pred)
print("R squared Error :",error_score)
```

Visualize the actual price and predicted prices

```python
plt.scatter(y_train,pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actaul prices vs Predicted prices")
plt.show()
```

```python
#prediction on testing data 
pred=lin_reg_model.predict(x_test)
```

```python
# R squared Error
error_score=metrics.r2_score(y_test,pred)
print("R squared Error :",error_score)
```

```python
plt.scatter(y_test,pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actaul prices vs Predicted prices")
plt.show()
```

2.Lasso Regression
